## Economic Indicators Trends

## GDP (constant and current LCU) over time

```python
import matplotlib.pyplot as plt

# Plot GDP (constant and current LCU) over time
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['GDP (constant LCU)'], label='GDP (constant LCU)', color='blue')
plt.plot(data['Year'], data['GDP (current LCU)'], label='GDP (current LCU)', color='orange')
plt.xlabel('Year')
plt.ylabel('GDP (LCU)')
plt.title('GDP (constant and current LCU) over Time')
plt.legend()
plt.grid(True)
plt.show()
```

The GDP (both constant and current LCU) has shown a significant upward trend over the years. This reflects the overall economic growth in Nigeria.

## Inflation (GDP deflator and consumer prices) over time

```python
# Plot Inflation (GDP deflator and consumer prices) over time
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['Inflation, GDP deflator (annual %)'], label='Inflation (GDP deflator)', color='green')
plt.plot(data['Year'], data['Inflation, consumer prices (annual %)'], label='Inflation (consumer prices)', color='red')
plt.xlabel('Year')
plt.ylabel('Inflation (%)')
plt.title('Inflation (GDP deflator and consumer prices) over Time')
plt.legend()
plt.grid(True)
plt.show()
```

* Inflation (both GDP deflator and consumer prices) has fluctuated significantly over the years. There have been periods of high inflation, especially in the late 1980s and early 1990s.

* 1981 : Extremely high GDP deflator inflation (219.00%) and consumer prices inflation (20.81%) following the global oil price collapse.

* 1988-1989: High consumer prices inflation rates (54.51% and 50.47%), reflecting economic challenges and instability.

* 1992-1995: Persistent high inflation, with GDP deflator peaking at 75.40% in 1995 and consumer prices inflation reaching 72.84%, likely due to structural adjustment programs and economic reforms.

## Petrol Price over time

```python
# Plot Petrol Price over time
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['Petrol Price (Naira)'], label='Petrol Price (Naira)', color='purple')
plt.xlabel('Year')
plt.ylabel('Petrol Price (Naira)')
plt.title('Petrol Price over Time')
plt.legend()
plt.grid(True)
plt.show()
```

The petrol price in Naira has seen a steady increase, particularly from the late 1990s onwards.
The most substantial rise appears around the 2000s, correlating with global oil price changes and domestic policy adjustments.

```python
# Analyze significant changes in petrol prices
significant_petrol_price_changes = data[['Year', 'Petrol Price (Naira)']].copy()
significant_petrol_price_changes['Change (%)'] = significant_petrol_price_changes['Petrol Price (Naira)'].pct_change() * 100

# Identify years with significant changes in petrol prices
high_petrol_price_changes = significant_petrol_price_changes[significant_petrol_price_changes['Change (%)'].abs() > 50]

high_petrol_price_changes
```

## GDP growth rate over time

```python
# Plot GDP growth rate over time
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['GDP growth (annual %)'], label='GDP growth (annual %)', color='blue')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.title('GDP Growth Rate over Time')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Identify key periods in GDP growth rate
significant_gdp_growth = data[['Year', 'GDP growth (annual %)']].copy()
significant_gdp_growth = significant_gdp_growth[(significant_gdp_growth['GDP growth (annual %)'] > 10) | (significant_gdp_growth['GDP growth (annual %)'] < -10)]

significant_gdp_growth
```

```python
# Plotting Total Reserves over the years
plt.figure(figsize=(8, 4))

plt.plot(data['Year'], data['Total reserves (includes gold, current US$)'], label='Total Reserves', color='brown')
plt.xlabel('Year')
plt.ylabel('Total Reserves (US$)')
plt.title('Total Reserves Over the Years')
plt.legend()

plt.tight_layout()
plt.show()
```

The trend shows a significant increase in total reserves over the years, indicating better accumulation of foreign exchange and gold reserves.

## Population trends over time

```python
# Plot population trends over time
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['Population, total'], label='Total Population', color='blue')
plt.plot(data['Year'], data['Population, female'], label='Female Population', color='pink')
plt.plot(data['Year'], data['Population, male'], label='Male Population', color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trends over Time')
plt.legend()
plt.grid(True)
plt.show()
```

The total population, along with female and male populations, shows a steady growth trend from 1960 onwards.
Growth rates have been relatively stable, around 2% annually for both male and female populations.

## Monetary Indicators Trends

```python
# Plotting Narrow Money, Money Supply M3, and Credit to Private Sector over the years
plt.figure(figsize=(10, 10))

# Narrow Money
plt.subplot(3, 1, 1)
plt.plot(data['Year'], data['Narrow Money'], label='Narrow Money')
plt.xlabel('Year')
plt.ylabel('Narrow Money')
plt.title('Narrow Money Over the Years')
plt.legend()

# Money Supply M3
plt.subplot(3, 1, 2)
plt.plot(data['Year'], data['Money Supply M3'], label='Money Supply M3', color='purple')
plt.xlabel('Year')
plt.ylabel('Money Supply M3')
plt.title('Money Supply M3 Over the Years')
plt.legend()

# Credit to Private Sector
plt.subplot(3, 1, 3)
plt.plot(data['Year'], data['Credit to Private Sector'], label='Credit to Private Sector', color='green')
plt.xlabel('Year')
plt.ylabel('Credit to Private Sector')
plt.title('Credit to Private Sector Over the Years')
plt.legend()

plt.tight_layout()
plt.show()
```

* Narrow Money: There is a noticeable upward trend in narrow money, reflecting an increase in the amount of money in circulation in the economy.
* Money Supply (M3): Similar to narrow money, money supply (M3) has also been increasing over the years.
* Credit to Private Sector: There is a steady increase in credit extended to the private sector, indicating growth in private sector activities and investments.

## Crude Oil: Price and Production

```python
# Plotting Crude Oil Price and Cumulative Crude Oil Production over the years
plt.figure(figsize=(14, 6))

# Crude Oil Price
plt.subplot(1, 2, 1)
plt.plot(data['Year'], data['Crude oil price(per barrel in $)'], label='Crude Oil Price', color='red')
plt.xlabel('Year')
plt.ylabel('Crude Oil Price (per barrel in $)')
plt.title('Crude Oil Price Over the Years')
plt.legend()

# Cumulative Crude Oil Production
plt.subplot(1, 2, 2)
plt.plot(data['Year'], data['Cumulative crude oil production up to and including year'], label='Cumulative Crude Oil Production', color='green')
plt.xlabel('Year')
plt.ylabel('Cumulative Crude Oil Production')
plt.title('Cumulative Crude Oil Production Over the Years')
plt.legend()

plt.tight_layout()
plt.show()
```

* Crude Oil Price: The price of crude oil has experienced considerable volatility, with significant peaks and troughs over the years.
* Cumulative Crude Oil Production: There is a clear upward trend in cumulative crude oil production, indicating a continuous increase in oil extraction activities over time.

Increased production helps generate revenue, but price volatility can make it difficult to achieve financial stability and consistent economic growth.
