**Variability**

*Variability refers to the extent of differences or fluctuations within a set of data points. It measures how much individual values deviate from the central tendency (such as the mean or median).*

Higher variability implies greater uncertainty, which may affect investment decisions.

Lower variability indicates stable processes.

```python

#Defining indicator
indicators = ['Agriculture', 'CropProduction', 'Livestock', 'Forestry', 'Fishing',
           'Industry', 'MiningAndQuarrying', 'CrudePetroleumAndNaturalGas',
           'CoalMining','Services']
# Calculate standard deviation for each sector
sector_std = df[indicators].std().sort_values(ascending=False)
print("Standard Deviation of Each Sector:\n", sector_std)

# Plot the standard deviation
plt.figure(figsize=(7, 7))
sns.barplot(x=sector_std.index, y=sector_std.values, hue=sector_std.index,palette='plasma')
plt.title('Standard Deviation of Sector')
plt.xlabel('Sector')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=90)  # Rotate sector names for better readability
plt.show()
```

**Variability Analysis of Economic Sectors**

1. **High Variability**:

   - **Services**:shows the  highest standard deviation (approximately 12736.28) indicates significant variability in this sector.

   - **Agriculture**: The second highest standard deviation (approximately 5866.18) indicates significant variability in this sector.
   - **Crop Production**: Following closely, crop production has a standard deviation of around 5401.48.

2. **Moderate Variability**:
   - **Industry**: The industrial sector shows moderate variability with a standard deviation of approximately 2491.22.

3. **Low Variability**:
   - **Crude Petroleum and Natural Gas**: This sector exhibits lower variability, with a standard deviation of about 1458.80.
   - **Mining and Quarrying**: Similar to petroleum and gas, mining and quarrying have a standard deviation of approximately 1448.67.
   - **Livestock**: Livestock shows relatively low variability, with a standard deviation of around 310.52.
   - **Fishing**: The fishing sector has even less variability, with a standard deviation of about 117.02.
   - **Forestry**: Forestry has minimal dispersion, with a standard deviation of approximately 41.89.
   - **Coal Mining**: Coal mining has the lowest standard deviation among all sectors, with a value of approximately 7.43.


**correlation Analysis**

Correlation Matrix: The values range from -1 to 1, where values close to 1 indicate a strong positive correlation, values close to -1 indicate a strong negative correlation, and values around 0 indicate no correlation.

 - High positive correlation, it means that as one variable increases, the other tends to increase as well.

 - High negative correlation indicates that as one variable increases, the other decreases.

```python
gdp_column = 'GDP at 2010 Constant Market Prices'

# Calculate the correlation matrix
correlation_matrix = df[[gdp_column] + indicators].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Factors and GDP')
plt.show()
```

**Time Series Analysis**

```python
# Plotting the trends over time for key sectors
key_sectors = ['Agriculture', 'Industry', 'Services', 'GDP at 2010 Constant Market Prices']
plt.figure(figsize=(14, 10))

for sector in key_sectors:
    plt.plot(df['Time'], df[sector], label=sector)

plt.xlabel('Time')
plt.ylabel('Value (in million Naira)')
plt.title('Trends over Time for Key Sectors')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.show()
```

-  Agriculture (blue line): Gradual, moderate, but steady growth.
-  Industry (orange line): Relatively flat with minor fluctuations, slight increase in the early 2000s.
- Services (green line): Substantial growth, especially from the early 2000s, closely following the overall GDP trend.

**Trade VS GDP Index**

```python

years = df['Time']
trade_data = df['Trade']
gdp_index = df['GDP at 2010 Constant Market Prices']

# Plotting trade vs GDP index
plt.figure(figsize=(12, 6))

plt.plot(years, trade_data, 'bo-', label='Trade')  # Blue solid line with circle markers
plt.plot(years, gdp_index, 'rs--', label='GDP Index')  # Red dashed line with square markers

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trade vs GDP Index Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyze the relative contributions of broad sectors
sector_df = df[['Agriculture', 'Industry', 'Services']].sum().to_frame('GDP')
sector_df['Percentage'] = (sector_df['GDP'] / sector_df['GDP'].sum()) * 100
sector_df = sector_df.sort_values('Percentage', ascending=False)

plt.figure(figsize=(8, 6))
sector_df['Percentage'].plot(kind='bar')
plt.title('Relative Contributions of Broad Sectors')
plt.xlabel('Sector')
plt.ylabel('Percentage of GDP')
plt.show()
```
