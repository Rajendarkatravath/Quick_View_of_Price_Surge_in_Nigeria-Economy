```python
import pandas as pd

file_path = '1980_onwards.csv'
data = pd.read_csv(file_path)

data.head()
```

## Population Growth Over Time

```python
import matplotlib.pyplot as plt

# Population Growth Over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Year'], data['Population'], marker='o')
plt.title('Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

The population of Nigeria has shown a consistent and significant increase from 1980 to 2023.

## Agricultural Production Over Time

```python
# Agricultural Production Over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Year'], data['Crop production'], label='Crop production')
plt.plot(data['Year'], data['Livestock'], label='Livestock')
plt.plot(data['Year'], data['Forestry'], label='Forestry')
plt.plot(data['Year'], data['Fishing'], label='Fishing')
plt.title('Agricultural Production Over Time')
plt.xlabel('Year')
plt.ylabel('Production Volume')
plt.legend()
plt.grid(True)
plt.show()
```

* Crop Production: There is a noticeable upward trend in crop production, particularly from the mid-1990s onwards, indicating increased agricultural productivity or expanded agricultural activities.
* Livestock: Livestock production shows a more gradual increase with some fluctuations.

## Sector Values Over Time

```python
# Sector Values Over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Year'], data['Construction Sector Naira(million)'], label='Construction')
plt.plot(data['Year'], data['Information and Communication Sector (millions in naira)'], label='Info & Comm')
plt.plot(data['Year'], data['Real Estate (millions of naira)'], label='Real Estate')
plt.title('Sector Values Over Time')
plt.xlabel('Year')
plt.ylabel('Value in Naira (millions)')
plt.legend()
plt.grid(True)
plt.show()
```

* Construction Sector: The value of the construction sector has shown substantial growth, particularly post-2000.
* Information and Communication Sector: This sector has also grown significantly, especially from the early 2000s.
* Real Estate: The real estate sector has shown a steady increase in value, mirroring trends in urbanization and real estate development.

## Value Added by Sectors (% of GDP) Over Time

```python
# Value Added by Sectors (% of GDP) Over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Year'], data['Manufacturing, value added (% of GDP)'], label='Manufacturing')
plt.plot(data['Year'], data['Agriculture, forestry, and fishing, value added (% of GDP)'], label='Agriculture')
plt.plot(data['Year'], data['Industry (including construction), value added (% of GDP)'], label='Industry')
plt.title('Value Added by Sectors (% of GDP) Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of GDP')
plt.legend()
plt.grid(True)
plt.show()
```

* Manufacturing: The contribution of manufacturing to GDP has shown variability, with periods of increase and decrease.
* Agriculture: The value added by agriculture has increased as a percentage of GDP.
* Industry: The industry's contribution (including construction) ha a dip in 2010, but it continues to increase as the years go by.

## Lending Interest Rate Over Time

```python
# Lending Interest Rate Over Time
plt.figure(figsize=(8, 4))
plt.plot(data['Year'], data['Lending interest rate (%)'], marker='o')
plt.title('Lending Interest Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Interest Rate (%)')
plt.grid(True)
plt.show()
```

* The lending interest rate has exhibited considerable variation, with a marked increase in the late 1990s and early 2000s. This could reflect changes in monetary policy, inflation rates, and economic conditions.
