```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_fuel = pd.read_csv('/content/drive/MyDrive/quickview of price surge in nigeria/Monthly/avg_fuel_pr.csv')

# Display the first 5 rows
print(df_fuel.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df_fuel.info())
```

```python
import plotly.express as px

# Convert `Month_Year` to datetime
df_fuel['Month_Year'] = pd.to_datetime(df_fuel['Month_Year'], format='%d/%m/%Y')

# Create a line chart of `Price in Naira Per Litre` over time
fig = px.line(df_fuel, x='Month_Year', y='Price in Naira Per Litre', title='Price of Fuel in Nigeria Over Time (1985-2024)')

# Add x and y axis labels
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Price in Naira Per Litre')

# Display plot
fig.show()
```

The chart illustrates the price of fuel in Nigeria from 1985 to 2024. There are several key trends:

Relative Stability (1985-1993): Fuel prices remained very low and stable during this period.

Sharp Increases (1994, 1998, 2000s): Prices began to rise more noticeably, with a few sharp spikes in 1994 and 1998. This upward trend continued into the 2000s.


Continued Upward Trend (2010s): The price continued to increase throughout the 2010s, albeit at a more gradual pace compared to the earlier spikes.


Significant Jump (2023-2024): The most dramatic increase occurred in 2023 and continued into 2024, with prices more than tripling.
